```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
z = [1, 8, 27, 64]

plt.subplot (2,1,1) # 2-rows, 1-column,1st plot
plt.plot (x,y,c='r', label = "Line 1")
plt.title("1st subplot")
plt.legend(loc="best")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(2,1,2)# 2-rows, 1-columns, 2nd plot
plt.plot(x,y,c="g",label = "Line 2")
plt.title("2nd subplot")
plt.legend(loc = 'best')
plt.xlabel("X-axis")
plt.ylabel("y-axis")

plt.suptitle(" Subplot")
plt.show()
```
